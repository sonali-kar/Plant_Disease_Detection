```python
Blight_dir = os.path.join('data/train/Blight')
Common_rust_dir = os.path.join('data/train/Common_Rust')
Gray_leaf_spot_dir = os.path.join('data/train/Gray_Leaf_Spot')
Healthy_dir = os.path.join('data/train/Healthy')


print('Total training Blight images:', len(os.listdir(Blight_dir)))
print('Total training Common rust images:', len(os.listdir(Common_rust_dir)))
print('Total training Gray leaf spot images:', len(os.listdir(Gray_leaf_spot_dir)))
print('Total training Healthy images:', len(os.listdir(Healthy_dir)))
```

```python
Blight_files = os.listdir(Blight_dir)
print(Blight_files[:10])

CommonRust_files = os.listdir(Common_rust_dir)
print(CommonRust_files[:10])

Gray_leaf_spot_files = os.listdir(Gray_leaf_spot_dir)
print(Gray_leaf_spot_files[:10])

Healthy_files = os.listdir(Healthy_dir)
print(Healthy_files[:10])
```

```python
pic_index = 2

next_Blight = [os.path.join(Blight_dir, fname) 
                for fname in Blight_files[pic_index-2:pic_index]]
next_CommonRust = [os.path.join(Common_rust_dir, fname) 
                for fname in CommonRust_files[pic_index-2:pic_index]]
next_GrayLeafSpot = [os.path.join(Gray_leaf_spot_dir, fname) 
                for fname in Gray_leaf_spot_files[pic_index-2:pic_index]]
next_Healthy = [os.path.join(Healthy_dir, fname) 
                for fname in Healthy_files[pic_index-2:pic_index]]


for i, img_path in enumerate(next_Blight+next_CommonRust+next_GrayLeafSpot+next_Healthy):
    img = mpimg.imread(img_path)
    plt.imshow(img)
    plt.axis('Off')
    plt.show()
```

```python
#data preprocessing

TRAINING_DIR = "data/train"
training_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
VALIDATION_DIR = "data/test"
validation_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_generator = training_datagen.flow_from_directory(
    TRAINING_DIR,
    target_size=(150,150),
    class_mode='categorical')

validation_generator = validation_datagen.flow_from_directory(
    VALIDATION_DIR,
    target_size=(150,150),
    class_mode='categorical')
```
